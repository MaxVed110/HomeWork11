```python
import sympy
from sympy import *
import matplotlib.pyplot
```

```python
x = Symbol('x', real = True)
```

```python
#уравнение
y = x**2-6*abs(x)+8
y
```

```python
#корни
solve(y,x)
```

```python
#график
f = plot(y,(x,-10,10), color = 'indigo')
```

```python
#экстремумы
diff_pos = y.diff(x)
print(diff_pos)
#строим график производной и видим кусочно-постоянную функцию 
plot(diff_pos, (x,0.1,5))
plot(diff_pos, (x,-0.1,-5))
```

```python
#рассмотрим положительное направление оси x, получим первый экстремум
xp = Symbol('x', real=True, positive=True)  

yp = x2**2-6*abs(x2)+8
print(yp)

diff_posp = yp.diff(xp)  # дифференцируем y(x), нуль производной даст точку экстремума функции

print(diff_posp)

zerosp = solveset(diff_posp, xp)
print(zerosp)
```

```python
#рассмотрим отрицательное направление оси x, получим второй экстремум
xn = Symbol('x', real=True, positive=False)  

yn = xn**2-6*abs(xn)+8
print(yn)

diff_posn = yn.diff(xn)  # дифференцируем y(x), нуль производной даст точку экстремума функции

print(diff_posn)

zerosn = solveset(diff_posn, xn)
print(zerosn)

##значения функции в экстремумах:
x3 = 3
y3 = x3**2-6*abs(x3)+8

x4 = -3
y4 = x4**2-6*abs(x4)+8

#в точке, где производная не существует
x5 = 0
y5 = x5**2-6*abs(x5)+8
print(y3)
print(y4)
print(y5)
```

```python
#таким образом, функция убывает на лучах (-оо, -3] и [0, 3]
#возрастает [-3, 0] и [3, +оо)
#значения функции в экстремумах отрицательные, следовательно функция больше 0: (-oo, -4] ⋃ [-2, 2] ⋃ [4, +oo)
#меньше 0: [-4, -2] ⋃ [2,4]
```
